```python
# CROSS VALIDATION TECHNIQUES AND HYPER PARAMETER TUNING 
```

```python
# CROSS VALIDATION SCORE
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(),x,y,cv=5)
```

```python
cross_val_score(SVC(C=10,gamma='auto',kernel='linear'),x,y,cv=5)
```

```python
# GRID SEARCH CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs = GridSearchCV(SVC(),
                 {
                     'C':[1,4,5,6,7,8,10],
                     'kernel':['rbf','linear'],
                     'gamma':['auto','scale']
                      
                 },cv=4,return_train_score=True)

gs.fit(x,y)
gs.cv_results_
```

```python
df1=pd.DataFrame(gs.cv_results_)   # To convert it into Data Frame
df1
```

```python
df1[['param_C','param_gamma','param_kernel','mean_test_score']]
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
# Randomized Search CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs = RandomizedSearchCV(SVC(),
                 {
                     'C':[1,4,5,6,7,8,10],
                     'kernel':['rbf','linear'],
                     'gamma':['auto','scale']
                      
                 },cv=4,return_train_score=True,n_iter=4)

rs.fit(x,y)
rs.cv_results_
```

```python
df2 = pd.DataFrame(rs.cv_results_)
df2
```

```python
df2[['param_C','param_gamma','param_kernel','mean_test_score']]
```

```python
rs.best_params_
```

```python
rs.best_score_
```

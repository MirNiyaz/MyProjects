```python
df["Branch"]=df.Branch.bfill()        #All The Null Values of Branch Takes its Succeeding Value.
```

```python
df
```

```python
df["City"]=df.City.ffill()             #All The Null Value of City Takes its Preceding value
```

```python
df
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df
```

```python
v=['Quantity']
```

```python
imputer=SimpleImputer(missing_values=np.NaN,strategy='mean')     #To Fill The Null Value taking Mean Using SimpleImputer
df[v]=imputer.fit_transform(df[v])                                      
```

```python
df
```

```python
v=['Date']
```

```python
imputer=SimpleImputer(missing_values=np.NaN,strategy='most_frequent')     #To Fill The Null Value Taking Frequent_Value Using SimpleImputer
df[v]=imputer.fit_transform(df[v])                                      
```
